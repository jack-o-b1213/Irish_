```python
import pandas as pd
import numpy
import os
```

```python
prices_df = pd.read_csv('data/housingPrices_cleansed.csv')
```

```python
prices_df
```

```python
def set_region(x):
    x = x.strip()
    if x == 'Co. Dublin':
        return 'Dublin'
    elif x == 'Co. Cork':
        return 'Cork'
    elif x == 'Co. Galway':
        return 'Galway'
    elif x == 'Co. Limerick':
        return 'Limerick'
    elif x == 'Co. Waterford':
        return 'Waterford'
    else: 
        return 'Other areas'
```

```python
prices_df['area'] = prices_df['county'].apply(lambda x: set_region(x))
del prices_df['index']
prices_df[prices_df.columns[0:]] = prices_df[prices_df.columns[0:]].replace('[\€,]', '', regex=True)
prices_df.price = prices_df.price.astype(float)
prices_df
```

```python
# prices_df[prices_df.columns[0:]] = prices_df[prices_df.columns[0:]].replace('[\€,]', '', regex=True)

cork_prices_df = prices_df.loc[prices_df["area"] == 'Cork']
dublin_prices_df = prices_df.loc[prices_df["area"] == 'Dublin']
galway_prices_df = prices_df.loc[prices_df["area"] == 'Galway']
limerick_prices_df = prices_df.loc[prices_df["area"] == 'Limerick']
waterford_prices_df = prices_df.loc[prices_df["area"] == 'Waterford']
other_prices_df = prices_df.loc[prices_df["area"] == 'Other areas']
```

```python
print("Cork avg. house price: €{:,.2f}".format(cork_prices_df['price'].mean()))
print("Dublin avg. house price: €{:,.2f}".format(dublin_prices_df['price'].mean()))
print("Galway avg. house price: €{:,.2f}".format(galway_prices_df['price'].mean()))
print("Limerick avg. house price: €{:,.2f}".format(limerick_prices_df['price'].mean()))
print("Waterford avg. house price: €{:,.2f}".format(waterford_prices_df['price'].mean()))
print("Other Areas avg. house price: €{:,.2f}".format(other_prices_df['price'].mean()))
print("National avg. house price: €{:,.2f}".format(prices_df['price'].mean()))
```

```python
prices_2021 = [cork_prices_df['price'].mean(), dublin_prices_df['price'].mean(), galway_prices_df['price'].mean(), limerick_prices_df['price'].mean(), waterford_prices_df['price'].mean(), other_prices_df['price'].mean(), prices_df['price'].mean()]
prices_2006Q1 = [303843, 431812.50, 308028.50, 253286.50, 261331.50, 274095, 321588.50]
```

```python
prices_2021
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(7)
# plot data in grouped manner of bar type
width = 0.2
plt.bar(x-0.1, prices_2021, width)
plt.bar(x+0.1, prices_2006Q1, width, color='cyan')
plt.xticks(x, ['Cork', 'Dublin', 'Galway', 'Limerick', 'Waterford', 'Other', 'National'])
plt.xlabel("Area")
plt.ylabel("Avg. Price (€)")
plt.legend(["2021", "2006 Q1"])
plt.show()
```
